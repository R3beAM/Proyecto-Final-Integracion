##Tipo de problema de Machine Learning

###Clasificación multiclase con NLP

El enfoque seleccionado es clasificación multiclase, porque el modelo no busca predecir un valor numérico continuo, como ocurriría en una regresión, sino asignar cada reseña a una de varias categorías posibles.

En este caso, las clases son:

* Sentimiento positivo
* Sentimiento neutral
* Sentimiento negativo

Además, el proyecto utiliza NLP porque la variable principal de entrada es texto no estructurado, específicamente las reseñas escritas por los clientes. Para que el modelo pueda procesar ese texto, será necesario aplicar técnicas de limpieza, transformación y vectorización, como TF-IDF o Bag of Words.

###¿Por qué se eligió este enfoque?

Se eligió este enfoque porque el objetivo del proyecto es identificar automáticamente la polaridad emocional de una reseña, es decir, determinar si el comentario expresa una experiencia favorable, desfavorable o intermedia sobre un producto.

Las reseñas de clientes contienen información subjetiva, como opiniones, satisfacción, molestias, recomendaciones o inconformidades. Esta información no puede analizarse eficientemente solo con métodos manuales, especialmente cuando existen grandes volúmenes de comentarios.

Por esta razón, un modelo de Machine Learning supervisado puede aprender patrones presentes en los textos y asociarlos con una categoría de sentimiento. Por ejemplo, palabras como “excellent”, “love”, “great” o “recommend” pueden estar asociadas a comentarios positivos, mientras que expresiones como “bad”, “disappointed”, “poor quality” o “not working” pueden estar relacionadas con comentarios negativos.

Este enfoque permite automatizar el análisis de reseñas, identificar tendencias de satisfacción e insatisfacción, y generar información útil para la toma de decisiones.

###Variable que se desea predecir

La variable que se desea predecir es el **sentimiento de la reseña**.

Esta variable puede llamarse, por ejemplo:

*sentiment*

El modelo recibirá como entrada el texto de la reseña, por ejemplo:

*reviews.text*

Y deberá predecir si esa reseña corresponde a una opinión positiva, negativa o neutral.

###¿Qué representa el target?

El target representa la categoría de sentimiento asociada a cada reseña.

En este proyecto, el target puede construirse a partir de la calificación del cliente, por ejemplo usando la variable:

*reviews.rating*

Una forma posible de definir el target sería:

**Calificación del cliente	Sentimiento asignado**
- 1 o 2 estrellas	Negativo
- 3 estrellas	Neutral
- 4 o 5 estrellas	Positivo

De esta forma, el modelo puede aprender a relacionar el contenido textual de la reseña con una etiqueta de sentimiento.

Por ejemplo:

Texto de reseña	Rating	Target
* “The product works perfectly and I love it.”	5	Positivo
* “It is okay, but nothing special.”	3	Neutral
* “The product stopped working after one week.”	1	Negativo



El target no representa directamente la calificación numérica del producto, sino una interpretación categórica del nivel de satisfacción expresado por el cliente.
